```python
import random

intent_responses = {
    "greeting": ["Hello! How can I assist you today?", "Hi there!"],
    "goodbye": ["Goodbye! Have a great day!", "Take care!"],
    "thanks": ["You're welcome!", "Any time!"],
    "default": ["I'm not sure I understand. Could you please clarify?"]
}

def recognize_intent(user_input):
    input_lower = user_input.lower()
    if any(word in input_lower for word in ['hello', 'hi', 'hey']):
        return "greeting"
    elif any(word in input_lower for word in ['bye', 'goodbye']):
        return "goodbye"
    elif any(word in input_lower for word in ['thank', 'thanks']):
        return "thanks"
    else:
        return "default"
```

```python
def chatbot_response(user_input):
    emotion = emotion_model.predict([user_input])[0]
    intent = recognize_intent(user_input)
    response = random.choice(intent_responses.get(intent, intent_responses["default"]))
    return f"[Detected Emotion: {emotion.capitalize()}] 🤖 {response}"
```

```python
import gradio as gr

iface = gr.Interface(
    fn=chatbot_response,
    inputs=gr.Textbox(lines=2, placeholder="Type your message here..."),
    outputs="text",
    title="🤖 Intelligent Customer Support Chatbot",
    description="Chatbot with real-time emotion detection and intelligent response."
)

iface.launch()
```

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import gradio as gr
import random

# Load dataset
df = pd.read_csv("Emotion_classify_Data[1].csv")
# Check the actual column names in the DataFrame
print(df.columns)
# Assuming the actual column names are 'Comment' and 'Emotion'
# Update the column selection accordingly
df = df[['Comment', 'Emotion']]  # Changed from 'comment' and 'emotion'

# Train emotion classifier
emotion_model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])
emotion_model.fit(df['Comment'], df['Emotion'])  # Changed from 'comment' and 'emotion'

# Intent-based replies
intent_responses = {
    "greeting": ["Hi! How can I help?", "Hello there!"],
    "goodbye": ["Bye! Come back soon.", "Goodbye!"],
    "thanks": ["You're welcome!", "Glad to assist!"],
    "default": ["Can you clarify that?"]
}

def recognize_intent(user_input):
    user_input = user_input.lower()
    if "hi" in user_input or "hello" in user_input:
        return "greeting"
    elif "bye" in user_input:
        return "goodbye"
    elif "thank" in user_input:
        return "thanks"
    return "default"

def chatbot_response(user_input):
    emotion = emotion_model.predict([user_input])[0]
    intent = recognize_intent(user_input)
    reply = random.choice(intent_responses.get(intent, intent_responses["default"]))
    return f"[Emotion: {emotion.capitalize()}] 🤖 {reply}"

# Gradio UI
iface = gr.Interface(
    fn=chatbot_response,
    inputs=gr.Textbox(lines=2, placeholder="Ask your question..."),
    outputs="text",
    title="🤖 AI Customer Support Bot",
    description="Chatbot with NLP + Emotion Detection for smart support."
)

iface.launch()
```
